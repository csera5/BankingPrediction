```python
#link to dataset: https://archive.ics.uci.edu/dataset/222/bank+marketing
'''
Input dataset from csv file
'''
import pandas as pd
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt

dataset = pd.read_csv('bank_dataset.csv', sep=';', header=0)
'''Save an untouched version of the dataset'''
dataset_original = dataset.copy()

'''Check that csv was copied properly'''
print(dataset_original)
```

```python
'''
Remove some unneeded features, including "contact", "day_of_week", "month", "duration"

Reasons:
duration: last contact duration, in seconds (numeric). Important note:  this attribute highly affects the output target (e.g., if duration=0 then y="no"). Yet, the duration is not known before a call is performed. Also, after the end of the call y is obviously known. Thus, this input should only be included for benchmark purposes and should be discarded if the intention is to have a realistic predictive model.
'''

dataset = dataset.drop(['contact', 'day_of_week', 'month', 'duration'], axis=1)

print(dataset)
```

```python
'''To make this a binary output, replace any yes or no answers in the target variable to be 0 or 1, this will be easier for ML later on'''
dataset['y'] = dataset['y'].replace({'no': 0, 'yes': 1})
```

```python
'''checking default vs target '''
dataset = dataset[dataset['default'] != 'unknown']

contingency_table = pd.crosstab(dataset['default'], dataset['y'])

print("contingency table for default vs target")
# Print the contingency table
print(contingency_table)

# Perform the chi-square test
chi2, p, _, _ = chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")

#28391 instances where both default and y were no 
#4197 instances where y was yes and default was no 
#3 instances where default was yes and y was no
#0 instances where both default and y were yes 
# conjecture is kind of supported to an extent but it isnt very valid since there weren't many people in default yes anyways 
# p value of 1 is bad, it means there is no significant association betwene the two values, we reject my conjecture 

#convert to 0 and 1s 
dataset['default_numeric'] = dataset['default'].map({'no': 0, 'yes': 1})

#calculate the correlation coefficient
correlation_coefficient = dataset['default_numeric'].corr(dataset['y'])
print(f"Correlation coefficient: {correlation_coefficient}")

#correlation coefficient is 0 so defintely no association 
contingency_table.plot(kind='bar', stacked=False)
plt.xticks(rotation=0)
plt.xlabel('Default')
plt.ylabel('Count')
plt.title('Subscription to Term Deposit by Default Status')
plt.legend(title='Term Deposit', labels=['No', 'Yes'], loc='upper right')
plt.show()
```

```python
'''checking personal loan vs target '''
dataset = dataset[dataset['loan'] != 'unknown']

contingency_table_loan = pd.crosstab(dataset['loan'], dataset['y'])

print("contingency table for loan vs target")
# Print the contingency table
print(contingency_table_loan)

# Perform the chi-square test
chi2loan, ploan, _, _ = chi2_contingency(contingency_table_loan)
print(f"Chi-square statistic: {chi2loan}")
print(f"P-value: {ploan}")

#23347 instances where both loan and y were no 
#3478 instances where y was yes and loan was no 
#4353 instances where loan was yes and y was no
#623 instances where both loan and y were yes 
#of the people who have a personal loan only 12.5% subscribed to a term deposit - but this is not strongly supported by other stats (p val and chi^2)
# p value is greater than 0.05 so there is no significant association between the two values, we reject my conjecture 

#convert to 0 and 1s 
dataset['loan_numeric'] = dataset['loan'].map({'no': 0, 'yes': 1})

#calculate the correlation coefficient
correlation_coefficient_loan = dataset['loan_numeric'].corr(dataset['y'])
print(f"Correlation coefficient: {correlation_coefficient_loan}")

#correlation coefficient is 0 so no relationship

contingency_table_loan.plot(kind='bar', stacked=False)
plt.xticks(rotation=0)
plt.xlabel('Personal Loan')
plt.ylabel('Count')
plt.title('Subscription to Term Deposit by Personal Loan Status')
plt.legend(title='Term Deposit', labels=['No', 'Yes'], loc='upper right')
plt.show()  
```

```python
'''Checking if a model would be more or less accurate if the only features are debt related values'''
'''prepping for model and splitting features vs target values'''
dataset_model = dataset_original.copy()

binary_columns = ['default', 'housing', 'loan']
dataset_model[binary_columns] = dataset_model[binary_columns].replace({'no': 0, 'yes': 1})
dataset_model = dataset_model.replace('unknown', pd.NA).dropna()
dataset_model['y'] = dataset_model['y'].replace({'no': 0, 'yes': 1})

y = dataset_model['y']
X = dataset_model[binary_columns]

print(X)
```

```python
'''Splitting into testing and training data, this is for the model that will only use the debt related features'''
from sklearn import model_selection, linear_model, feature_selection, metrics
from sklearn.metrics import accuracy_score

# Spilt data into 80% training and 20% rest
X_train,X_,y_train,y_ = model_selection.train_test_split(X,
                                                         y,
                                                         test_size=0.2,
                                                         random_state=123)

# Split the test sets into half validation and half testing (10% each of total)
# We have 80% training, 10% validation, 10% testing
X_val,X_test,y_val,y_test = model_selection.train_test_split(X_,
                                                             y_,
                                                             test_size=0.5,
                                                             random_state=42)
```

```python
'''random forest algorithm for model with only debt related features'''
# Run Random Forest
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(n_estimators=100)
rf_classifier.fit(X_train, y_train.values.ravel())

# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)
y_pred_val = rf_classifier.predict(X_val)

# Evaluate accuracy
val_accuracy = accuracy_score(y_val, y_pred_val)
print(f'Validation Accuracy using Random Forest: {val_accuracy:.4f}')
ran_accuracy = accuracy_score(y_test, y_pred)
print(f'Test Accuracy using Random Forest:       {ran_accuracy:.4f}')
```

```python
'''logistic regression algorithm for model with only debt related features'''
# Run logistic regression
from sklearn.linear_model import LogisticRegression

logistic_regression = LogisticRegression(max_iter=5000)
logistic_regression.fit(X_train, y_train.values.ravel())

# Make predictions on test set
y_pred = logistic_regression.predict(X_test)
y_pred_val = logistic_regression.predict(X_val)

# Evaluate accuracy
val_accuracy = accuracy_score(y_val, y_pred_val)
print(f'Validation Accuracy using logistic regression: {val_accuracy:.4f}')
knn_accuracy = accuracy_score(y_test, y_pred)
print(f'Test Accuracy using logistic regression:       {knn_accuracy:.4f}')
```

Since the model using only debt related features had poor accuracy we decided to not move forward with this as our final model, instead we will continue to move with our afforementioned features that include demographic, debt related, and social economic factors. 

Now we will continue to develop a model using 3 different algorithms, we will choose our final model to be the one with the highest accuracy

```python
# Feature engineering
# replace columns with binary values with numerical values
final_model = dataset_original.copy()

binary_columns_c = ['default', 'housing', 'loan']
final_model[binary_columns_c] = final_model[binary_columns_c].replace({'no': 0, 'yes': 2, 'unknown': 1})
final_model['poutcome'] = final_model['poutcome'].replace({'failure': 0, 'success': 1, 'nonexistent': -99})
final_model['y'] = final_model['y'].replace({'no': 0, 'yes': 1})

# order job from lowest pay to highest pay in numerical values
final_model['job'] = final_model['job'].replace({'unknown': 0, 'student': 1, 'unemployed': 2, 'housemaid': 3, 'retired': 4, 'blue-collar': 5, 'admin.': 6, 'services': 7, 'technician': 8, 'self-employed': 9, 'entrepreneur': 10, "management": 11})

# order education from lowest to highest in numerical values
final_model['education'] = final_model['education'].replace({'unknown': 0, 'basic.4y': 1,'basic.6y': 2, 'basic.9y': 3, 'high.school': 4, 'illiterate': 5, 'professional.course': 6, 'university.degree': 7})

# categorize martial status
final_model['marital'] = final_model['marital'].replace({'unknown': 0, 'single': 1, 'married': 2, 'divorced': 3})

y_c = final_model['y']
X_c = final_model.drop(['y', 'duration', 'month', 'day_of_week', 'contact', 'age', 'job', 'housing', 'loan'], axis=1)

print(X_c)
```

```python
import matplotlib.pylab as py
from sklearn import model_selection, linear_model, feature_selection, metrics
from sklearn.metrics import accuracy_score

# Spilt data into 80% training and 20% rest
X_train_c,X_c,y_train_c,y_c = model_selection.train_test_split(X_c,
                                                         y_c,
                                                         test_size=0.2,
                                                         random_state=123)

# Split the test sets into half validation and half testing (10% each of total)
# We have 80% training, 10% validation, 10% testing
X_val_c,X_test_c,y_val_c,y_test_c = model_selection.train_test_split(X_c,
                                                             y_c,
                                                             test_size=0.5,
                                                             random_state=42)
```

```python
# Run Random Forest
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(n_estimators=100)
rf_classifier.fit(X_train_c, y_train_c.values.ravel())

# Make predictions on the test set
y_pred_c = rf_classifier.predict(X_test_c)
y_pred_val_c = rf_classifier.predict(X_val_c)

# Evaluate accuracy
val_accuracy_rf = accuracy_score(y_val_c, y_pred_val_c)
print(f'Validation Accuracy using Random Forest: {val_accuracy_rf:.4f}')
ran_accuracy_rf = accuracy_score(y_test_c, y_pred_c)
print(f'Test Accuracy using Random Forest:       {ran_accuracy_rf:.4f}')

cm_rf = confusion_matrix(y_test_c, y_pred_c)
disp_rf = ConfusionMatrixDisplay(confusion_matrix=cm_rf,
                               display_labels=rf_classifier.classes_)
disp_rf.plot()
disp_rf.ax_.set_title('Confusion Matrix for Random Forest')
```
